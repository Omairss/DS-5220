QUADRATIC LOSS

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
delta = 1
huber = lambda e, delta: ((e/2)**2) if np.abs(e) <= delta else ((delta * np.abs(e)) - ((delta**2)/2))
```

## 4A

```python
ax = plt.subplot(111)
t1 = np.arange(-2.0, 2.0, 0.01)
t2 = [huber(x, delta) for x in t1]

plt.plot(t1, t1**2, label="Quadratic Loss")
plt.plot(t1, np.abs(t1), label="Mean Absolute Error")
plt.plot(t1, t2, label="Huber Error")




leg = plt.legend(loc='best', ncol=2, mode="expand", shadow=False, fancybox=False)
leg.get_frame().set_alpha(0.5)


plt.show()
```

```python
def err_calc(y, y_pred):
    
    mean_error = []
    quadratic_error = []
    huber_loss = []
    for i in range(0,50):

        mean_error.append(y[i] - y_pred[i])
        quadratic_error.append((y[i] - y_pred[i])*2)
        huber_loss.append(huber(y[i] - y_pred[i], 3))
    
    return mean_error, quadratic_error, huber_loss
```

```python
t0, t1 = 2, 1
hypothesis = lambda x: x*t1+t0
y_pred = [hypothesis(i) for i in x]
error = err_calc(y, y_pred)
```

```python
mean_abs_error = sum(error[0])/50
quadratic_error = sum(error[1])/50
huber_loss = sum(error[2])/50
```

## Gradient Descent - Mean Absolute Error

```python
rate = 0.1 # Learning rate
precision = 0.000001 #This tells us when to stop the algorithm
previous_step_size = 1 #
max_iters = 5000 # maximum number of iterations
iters = 0
cur_t0, cur_t1 = 2, 1

while previous_step_size > precision and iters < max_iters:

    prev_t0 = cur_t0
    prev_t1 = cur_t1

    hypothesis = lambda x: x*cur_t1+cur_t0

    y_pred = [hypothesis(i) for i in x]
    error = (sum(y) - sum(y_pred)) / 50
    
    #Mean abs error
    if error > 0:
        delta0 = 1
    elif error < 0:
        delta0 = -1
    else:
        delta0 = 0 
        
    delta1 = 0
        
    cur_t0 = cur_t0 + rate * delta0
    cur_t1 = cur_t1 + rate * delta1
    
    previous_step_size_t0 = abs(cur_t0 - prev_t0) 
    previous_step_size_t1 = abs(cur_t1 - prev_t1) 

    iters += 1

    print("Iteration",iters,"\nT0 value is",cur_t0)
    print("Iteration",iters,"\nT1 value is",cur_t1)
    print("Error:" + str(error) + '\n')



print("The local minimum occurs at", cur_t0)
```

```python
def RidgeGradientDescent(x, y, alpha, iters, L):
    
    #x=np.matrix(x)
    intercept = np.ones(len(x))
    x = np.append(intercept,x)
    
    x = np.reshape(x,(50,2))
    y = np.matrix(y).transpose()
    m, n = np.shape(x)
    beta = np.matrix(np.ones(n)).transpose()
    XT = x.transpose()
    for i in range(0, iters):
        y_hat = np.dot(x, beta)
        residuals = y_hat - y
        MSE = (residuals.transpose()*residuals)/len(x)
        print ("iteration:", i, "MSE:", MSE)
        
        ols_gradient = np.dot(XT, residuals) / m
        beta = beta + alpha * (ols_gradient + (L/m)*beta)
    return beta
```
