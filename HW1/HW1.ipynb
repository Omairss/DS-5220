## Gradient Descent - Mean Absolute Error

```python
sim = lambda x: (3*random.uniform(-2, 2)*x)+2+random.gauss(0,4)
x, y = [x for x in range(0,50)],[sim(x) for x in range(0,50)]
```

```python
rate = 1 # Learning rate
precision = 0.01 #This tells us when to stop the algorithm
previous_step_size = 1 #
max_iters = 5000 # maximum number of iterations
iters = 0
cur_t0, cur_t1 = 5, 8

while previous_step_size > precision and iters < max_iters:

    prev_t0 = cur_t0
    prev_t1 = cur_t1

    hypothesis = lambda x: x*cur_t1+cur_t0

    y_pred = [hypothesis(i) for i in x]
    error = (sum(y) - sum(y_pred)) / 50
    
    #Mean abs error
    if error > 0:
        delta0 = 1
    elif error < 0:
        delta0 = -1
    else:
        delta0 = 0 
        
    delta1 = 0
        
    cur_t0 = cur_t0 + rate * delta0
    cur_t1 = cur_t1 + rate * delta1
    
    previous_step_size_t0 = abs(cur_t0 - prev_t0) 
    previous_step_size_t1 = abs(cur_t1 - prev_t1) 

    iters += 1

    print("Iteration",iters,"\nT0 value is",cur_t0)
    print("Iteration",iters,"\nT1 value is",cur_t1)
    print("Error:" + str(error) + '\n')



print("The local minimum occurs at", cur_t0)
```

```python
from numpy import array, dot, transpose
from numpy.linalg import inv

def analytical_linear_regression(x_train, y_train):
    
    X = np.array(x_train)
    ones = np.ones(len(X))
    X = np.column_stack((ones,X))
    y = np.array(y_train)
    
    Xt = transpose(X)
    product = dot(Xt, X)
    theInverse = inv(product)
    w = dot(dot(theInverse, Xt), y)
    
    return w
```

```python
def RidgeGradientDescent(x, y, alpha, iters, L):
    
    #x=np.matrix(x)
    intercept = np.ones(len(x))
    x = np.append(intercept,x)
    
    x = np.reshape(x,(50,2))
    y = np.matrix(y).transpose()
    m, n = np.shape(x)
    beta = np.matrix(np.ones(n)).transpose()
    XT = x.transpose()
    for i in range(0, iters):
        y_hat = np.dot(x, beta)
        residuals = y_hat - y
        MSE = (residuals.transpose()*residuals)/len(x)
        print ("iteration:", i, "MSE:", MSE)
        
        ols_gradient = np.dot(XT, residuals) / m
        beta = beta + alpha * (ols_gradient + (L/m)*beta)
    return beta
```

```python
x = np.random.uniform(-2, 2, size=50)
y = 3*x+2+np.random.normal(0,4,size=50)
plt.scatter(x, y, label="Quadratic Loss", s=10)
```

```python
def quadratic_loss(cur_t0, cur_t1):

    rate = 0.01 # Learning rate
    precision = 0.0000001 #This tells us when to stop the algorithm
    previous_step_size_t1 = 1 #
    max_iters = 5000 # maximum number of iterations
    iters = 0

    while previous_step_size_t1 > precision and iters < max_iters:

        prev_t0 = cur_t0
        prev_t1 = cur_t1

        hypothesis = lambda x: x*cur_t1+cur_t0

        y_pred = [hypothesis(i) for i in x]
        error = sum(np.array(y) - np.array(y_pred))

        #Quadratic error
        delta0 = error 

        delta1 = error * sum(x)
        #================
        
        cur_t0 = cur_t0 + (rate * delta0)
        cur_t1 = cur_t1 + (rate * delta1)

        previous_step_size_t0 = abs(cur_t0 - prev_t0) 
        previous_step_size_t1 = abs(cur_t1 - prev_t1) 

        iters += 1

        print("Iteration",iters,"\nT0 value is",cur_t0)
        print("Iteration",iters,"\nT1 value is",cur_t1)
        print("Error:" + str(error) + '\n')



    print("The local minimum occurs at", cur_t1)
    
    bf = lambda x: cur_t1*x+cur_t0
    plt.plot([x for x in range(0,500)],[bf(x) for x in range(0,500)], label="Quadratic Loss")
    plt.scatter(x, y, label="Quadratic Loss", s=10)
    plt.ylim(-10, 20)
    plt.xlim(-2,2)


    plt.show()


    
    return cur_t0, cur_t1, error
```
