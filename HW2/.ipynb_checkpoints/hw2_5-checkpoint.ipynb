### a

Analytical solution for least squares regression

```python
def linear_regression_analytical_sol(x, y): 
    x = np.array(x)
    y = np.array(y)
    xT = transpose(x)
    prod = dot(xT, x)
    inverse = inv(prod)
    tmp = dot(inverse, xT)
    theta = dot(tmp, y)
    return theta
```

```python
theta = linear_regression_analytical_sol(x, y)
theta
```

```python
reg = LinearRegression().fit(x, y)
reg.coef_
```

Analytical solution for ridge regression as a function of parameter lambda

```python
def linear_regression_analytical_ridge(x, y, l):
    # l is lambda
    x = np.array(x)
    y = np.array(y)
    xT = transpose(x)
    prod = dot(xT, x) + l
    inverse = inv(prod)
    tmp = dot(inverse, xT)
    theta = dot(tmp, y)
    return theta
```

```python
theta = linear_regression_analytical_ridge(x, y, 3)
theta
```

```python
def gradient_descent(loss_measure, x, y, theta, alpha, m, n, delta=2):
    """
    loss_type: string
    n: number of iterations
    """
    xT = x.transpose()
    for i in range(n):
        hypothesis = np.dot(x, theta)
        loss = hypothesis - y
        gradient = np.dot(xT, loss) / m
        theta = theta - alpha * gradient
    return theta
```

```python
def train_validate_test_split(df, train_percent=.38, validate_percent=.38, seed=None):
    np.random.seed(seed)
    perm = np.random.permutation(df.index)
    m = len(df.index)
    train_end = int(train_percent * m)
    validate_end = int(validate_percent * m) + train_end
    train = df.ix[perm[:train_end]]
    validate = df.ix[perm[train_end:validate_end]]
    test = df.ix[perm[validate_end:]]
    return train, validate, test
```

```python
train, validate, test = train_validate_test_split(df)
print(len(train))
print(len(validate))
print(len(test))
```

```python
# R2 and RMSE
def get_score(prediction, lables):    
    print('R2: {}'.format(r2_score(prediction, lables)))
    print('RMSE: {}'.format(np.sqrt(mean_squared_error(prediction, lables))))

# Accuracy Scores   
def train_test(estimator, x_trn, x_tst, y_trn, y_tst):
    prediction_train = estimator.predict(x_trn)
    print(estimator)
    print("Training Score = ")
    get_score(prediction_train, y_trn)
    prediction_test = estimator.predict(x_tst)
    print("Testing Score = ")
    get_score(prediction_test, y_tst)
```
